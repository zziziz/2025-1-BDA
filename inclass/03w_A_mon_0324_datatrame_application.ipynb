#데이터프레임수정

```python
df[['사망(명)', '부상(명)']].sum(axis=1)
```

```python
df2 = df.copy()
df2['재해자수'] = df[['사망(명)', '부상(명)']].sum(axis=1) # Changed '사명(명)' to '사망(명)'
df2
```

```python
df2.loc['총계']=df2.sum()
df2
```

```python
df[df2>210]
```

```python
df2[[True, False]*6] # remove the extra  + [True]
```

```python
df2[[True]*6 + [False]*6] # Changed the number of False values to 6
```

```python
df2['재해자수']>400
```

```python
df2[df2['재해자수']>400]
```
