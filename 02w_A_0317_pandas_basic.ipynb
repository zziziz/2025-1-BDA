시리즈 활용

```python
import numpy as np
import pandas as pd

s = ['희빈', '수영', '현수', '지연', '지민']
p = pd.Series([99, 84, 78, 99, 65], index=s)
p
```

```python
p.name = '수학성적'
p.index.name = '이름'
p
```

```python
p.sort_index()#가나다순으로정렬(오름차순)
```

```python
p.sort_index(ascending=False)#내림차순
```

```python
p.sort_values()
```

```python
p.sort_values(ascending=False)
```

```python
print(p.max())
print(p.min())
print(p.mean())
print(p.median())
print(p.count())
print(p.size) # Remove the parentheses after p.size
```

```python
p.describe()
```
